```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
year = "Resources/crimes_1985-2018.csv"
year_df = pd.read_csv(year)
year_df.head()
```

```python
#VehicleTheft_sum = Number of Motor Vehicle Thefts
#MVTact_sum = Automobiles Theft â€“ Autos - Actual


columns2 = ["County", "Year", "VehicleTheft_sum", "MVTact_sum"]
new_year = year_df[columns2].copy()

new_year.head()
```

```python
#filter for only years 2016-2018

mask = (new_year['Year'] >= 2016) & (new_year['Year'] <= 2018)
new_year = new_year.loc[mask]
new_year.head()
```

```python
cleaned_ca = new_year.groupby(['Year','County']).sum().reset_index()
cleaned_ca
```

```python
#Rename County names to drop County to match Carl's dataset for merge

cleaned_ca['Updated_County'] = cleaned_ca["County"].str[:-7]
cleaned_ca
```
